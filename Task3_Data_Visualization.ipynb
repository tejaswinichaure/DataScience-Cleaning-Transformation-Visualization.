```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:/Users/HP/Downloads/titanic.csv')  # Replace with your dataset path
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Explanation: The histogram uses sns.histplot with kde=True to show the kernel density estimate overlayed on the age distribution. This helps visualize where age clusters exist, which can reveal age demographics for the Titanic passengers.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

Explanation: The bar chart displays the survival rate across classes, using sns.barplot. By setting ci=None, confidence intervals are removed to simplify the view. This plot highlights how survival rates varied significantly by passenger class.

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Titanic Dataset Features')
plt.show()
```

Explanation: This heatmap uses sns.heatmap to visualize correlations between numerical features.
    annot=True adds correlation values inside the cells, and cmap='coolwarm' uses contrasting colors 
    to highlight positive and negative correlations.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survival Status (0 = Did Not Survive, 1 = Survived)')
plt.ylabel('Age')
plt.show()
```

Explanation: The violin plot is ideal for showing age distribution while comparing survival groups. This can reveal age groups with higher or lower survival rates, as well as the spread of ages within each group.
